```python
import pandas as pd
import numpy as np
```

```python
df_crimes = pd.read_csv("../data/crimes_clean.csv")
df_health = pd.read_csv("../data/health_clean.csv")
```

```python
df=df_crimes.merge(
    df_health,
    left_on="Community Area",
    right_on="community_area",
    how="left"
)
```

```python
df.drop(columns=["community_area"], inplace=True)
```

```python
print("\nCombined shape:", df.shape)
```

```python
calidad_vida = df["hardship_index"].isna().sum()

print(f"\nCrímenes sin datos de calidad de vida del barrio: {calidad_vida} ({calidad_vida/len(df)*100:.2f}%)")
```
