```python
import pandas as pd
import numpy as np
```

```python
df_health = pd.read_csv("../data/Chicago Health Atlas Data Download - Community areas.csv", sep=",", header=1)
print("Health shape dataset original:", df_health.shape)
```

```python
df_health.columns = [
    "area_type",              
    "neighborhood_name",      
    "community_area",         
    "unemployment_rate",      
    "hardship_index",         
    "median_income",          
    "people_of_color_pct",    
    "child_opportunity_index" 
]
```

```python
print("\nColumnas tras eliminar innecesarias:",df_health.columns.tolist())
```

```python
df_health.drop(columns=["area_type"], inplace=True)
```

```python
print("\nNulos en df_health:")
print(df_health.isnull().sum())
print(f"\nDuplicados en df_health: {df_health.duplicated().sum()}")
```

```python
print("\nHealth shape final:", df_health.shape)
print("\nPrimeras filas df_health:")
print(df_health.head())
```

```python
df_health.describe()
```

```python
print(df_health.isnull().sum())
```

```python
# Se categorizan según agrupaciones para simplificar:

df_health["child_opportunity_category"] = pd.cut(
    df_health["child_opportunity_index"],
    bins=[1, 20, 40, 60, 80, 101],
    labels=["Very Low", "Low", "Moderate", "High", "Very High"],
    right=False
)

df_health["unemployment_category"] = pd.cut(
    df_health["unemployment_rate"],
    bins=[0, 5, 10, float('inf')],
    labels=['Low', 'Moderate', 'High'],
    right=False
)

df_health["hardship_category"] = pd.cut(
    df_health["hardship_index"],
    bins=[1, 20, 40, 60, 80, 101],
    labels=["Very High", "High","Moderate","Low", "Very Low"],
    right=False
)

df_health['income_category'] = pd.cut(
    df_health['median_income'],
    bins=[0, 30000, 60000, 100000, 150000, float('inf')],
    labels=['Very Low', 'Low', 'Middle', 'Upper Middle', 'High']
)

df_health['people_of_color_category'] = pd.qcut(
    df_health['people_of_color_pct'],
    q=5,
    labels=['Very Low', 'Low', 'Moderate', 'High', 'Very High']
)
```

```python
df_health.to_csv("../data/health_clean.csv", index=False)
```
