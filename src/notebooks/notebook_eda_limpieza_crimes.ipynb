```python
import pandas as pd
import numpy as np

```

```python
df_crimes = pd.read_csv("../data/Chicago_Crimes_2014_to_2017.csv", sep=";")


print("Crimes shape dataset original:", df_crimes.shape)
```

```python
columns_delete = [
    "Unnamed: 0",
    "Case Number", 
    "IUCR",         
    "FBI Code",    
    "Beat",         
    "Block",        
    "Updated On",   
    "X Coordinate", 
    "Y Coordinate", 
       
]

df_crimes.drop(columns=columns_delete, inplace=True)
print("\nColumnas tras eliminar innecesarias:",df_crimes.columns.tolist())
```

```python
df_crimes["Date"] = pd.to_datetime(df_crimes["Date"], errors="coerce")
```

```python
df_crimes = df_crimes.dropna(subset=["Date"])
```

```python
def fix_coordinate(value):
    if pd.isna(value):
        return None
    value= str(value).strip()
    parts= value.split(".")
    if len(parts) <= 2:
        return float(value) 
    fixed = parts[0] + "." + "".join(parts[1:])
    return float(fixed)

df_crimes["Latitude"] = df_crimes["Latitude"].apply(fix_coordinate)
df_crimes["Longitude"] = df_crimes["Longitude"].apply(fix_coordinate)

print("\nTipo Latitude/Longitude tras conversión:", df_crimes["Latitude"].dtype)
print("Ejemplo Latitude:", df_crimes["Latitude"].dropna().head(3).values)
```

```python
df_crimes = df_crimes.dropna(subset=["Ward"])
```

```python
df_crimes["Community Area"] = df_crimes["Community Area"].astype("Int64")
```

```python
duplicados = df_crimes.duplicated().sum()
print("\nDuplicados encontrados:", duplicados)
df_crimes.drop_duplicates(inplace=True)

```

```python
df_crimes["Location Description"].fillna("Unknown", inplace=True)
print("\nNulos tras limpieza:",df_crimes)
print(df_crimes.isnull().sum())
```

```python
print("\nCrimes shape final:", df_crimes.shape)
```
